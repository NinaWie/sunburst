```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import Line2D, Polygon
from scipy import stats
from matplotlib.patches import Patch
import matplotlib
import pandas as pd
import os
matplotlib.rcParams.update({'font.size': 13})
```

```python
import json
with open(os.path.join("data", "dummy_data.json"), "r") as infile:
    dic = json.load(infile)
```

```python
category_3 = ["red", "cyan", "magenta", "yellow"]
category_2 = ["A", "B", "C", "D", "E"]
short_cat_2 = ["A", "B", "C", "D", "E"] # if the names of category_2 are to short to be displayed inside the plot, write short names for it in this list
cols = ['r', 'c', 'm', 'y'] # ["red", "green", "yellow", "purple"]
```

```python
def most_frequent(List): 
    return max(set(List), key = List.count) 

def get_xy(winkel_per_line, radius):
    height = np.cos(winkel_per_line) * radius
    width = np.sin(winkel_per_line) * radius
    return width, height

def get_polygon(winkel1, winkel2, r1,r2):
    x_1, y_1 = get_xy(winkel1, r1)
    x_2, y_2 = get_xy(winkel2, r1)
    x_3, y_3 = get_xy(winkel1, r2)
    x_4, y_4 = get_xy(winkel2, r2)
    return [[x_1, y_1],[x_3, y_3],[x_4, y_4],[x_2, y_2]]

def create_proxy(label):
    line = matplotlib.lines.Line2D([0], [0], linestyle='none', mfc='black',
                mec='none', marker=r'$\mathregular{{{}}}$'.format(label))
    return line
```

```python
fig, ax = plt.subplots(figsize=(15,15))

# ax.set_facecolor((0.1, 0.1, 0.1))

number_lines = len(dic.keys())+2
radius = 3

winkel_list = [(2*np.pi/number_lines) * winkel_ind for winkel_ind in range(number_lines)]
winkel_list = (np.roll(winkel_list,40)).tolist()
del(winkel_list[-2])

key_list = list(dic.keys())

prev_winkel = winkel_list[-1]
    
for winkel_ind, winkel in enumerate(winkel_list[:-1]):
    width,height = get_xy(winkel, radius)
    
    # dic stuff
    m_dic = dic[key_list[winkel_ind]]
    for cat_2 in m_dic.keys():
        if len(m_dic[cat_2])!=0:
            cat_3 = most_frequent(m_dic[cat_2])
            col = cols[category_3.index(cat_3)] # get colour corresponding
            r_start = (radius*0.95)*(category_2.index(cat_2)+1)/6
            ax.add_patch(Polygon(get_polygon(prev_winkel, winkel, r_start, r_start+(radius*0.95)/6), color=col))
    for k in range(5):
        r_start = (radius*0.95)*(k+1)/6
        ax.add_patch(Polygon(get_polygon(prev_winkel, winkel, r_start, r_start+(radius*0.95)/6), fill=False, color="grey"))
        
    ## Plotting stuff
    x = [0,width]
    y = [0,height]
    line = Line2D(x, y, c="grey")
    ax.add_line(line)
    ax.axis([-3.5, 3.5, -3.5, 3.5])
    if winkel_ind>0:
        ax.text(0.5*(width+ prev_pos[0]), 0.5*(height+prev_pos[1]), winkel_ind) # ,rotation=90*winkel/(2*np.pi))
    else:
        first_pos = (width, height)
    prev_pos = (width, height)
    prev_winkel = winkel


# last one: 
winkel = winkel_list[-1]
prev_pos = get_xy(winkel, radius)
x = [0,prev_pos[0]]
y = [0,prev_pos[1]]
line = Line2D(x, y, c="grey")
ax.add_line(line)
for k in range(5):
    r_start = (radius*0.95)*(k+1)/6
    poly = get_polygon(prev_winkel, winkel, r_start, r_start+(radius*0.95)/6)
    ax.add_patch(Polygon(poly, fill=False, color="grey"))
    # put text for legend
    center_box = 0.5*(np.asarray(poly[0])+np.asarray(poly[2]))
    ax.text(center_box[0], center_box[1], short_cat_2[k])

ax.text(0.5*(first_pos[0]+ prev_pos[0]), 0.5*(first_pos[1]+prev_pos[1]), 0)

# add white circle in middle
circle = plt.Circle((0, 0), 0.2*radius, color='white', fill=True, zorder=10)
ax.add_artist(circle)

legend_elements = [Patch(color=cols[i], label=category_3[i]) for i in range(len(category_3))]
l = plt.legend(handles=legend_elements, loc='upper left', frameon=False) 
l.set_zorder(0) 
plt.gca().add_artist(l)

# Create a legend for the industries
proxies = [create_proxy(item) for item in short_cat_2]
new_leg = ax.legend(proxies, category_2, numpoints=1, markerscale=2, loc='lower left', frameon=False)
plt.gca().add_artist(new_leg)

plt.axis("off")
# plt.savefig("assets/sunrays.png")
plt.show()
```
